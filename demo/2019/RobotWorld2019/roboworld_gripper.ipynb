# Indy Gripper - set wall

```python
 def open_gripper(prog):
    prog.add_digital_out(9, 1)
    prog.add_digital_out(8, 0)
    prog.add_wait(2)    
        
def close_gripper(prog):
    prog.add_digital_out(8, 1)
    prog.add_digital_out(9, 0)
    prog.add_wait(2)

def go_down(task_pos, i):
    one_step = 0.032
    task_pos[2] = task_pos[2] - one_step
    return task_pos

def get_handle_pos(pos1, pos2, i):
#     i~0:9
    pos = pos1 + ((pos2 - pos1)/9)*i  
    return pos
        
```

```python
def indy_gripper_setting_wall(vel=2):
    with open('teach-gripper-set.json') as json_file:
        tp_grip_set = json.load(json_file)

    prog = JsonProgramComponent(policy=0, resume_time=2)
    open_gripper(prog)
    prog.add_move_home()
    
    # First Long Wall
    # Pick     
    prog.add_joint_move_to(tp_grip_set['j_pre_pick_1'], vel)
    prog.add_task_move_to(tp_grip_set['t_pick_1'], vel)
    close_gripper(prog)
    prog.add_task_move_to(tp_grip_set['t_pickup_1'], vel)
    
    # Place     
    prog.add_move_home()
    prog.add_joint_move_to(tp_grip_set['j_pre_place_1'], vel)
    prog.add_task_move_to(tp_grip_set['t_place_1-1'], vel)
    prog.add_task_move_to(tp_grip_set['t_place_1-2'], vel)
    prog.add_task_move_to(tp_grip_set['t_place_1-3'], vel)
    open_gripper(prog)
    
    # Reposition    
    prog.add_task_move_to(tp_grip_set['t_reposition_1-1'], vel)
    prog.add_task_move_to(tp_grip_set['t_reposition_1-2'], vel)
    prog.add_task_move_to(tp_grip_set['t_reposition_1-3'], vel)
    prog.add_task_move_to(tp_grip_set['t_reposition_1-4'], vel)
    prog.add_move_home()
    
    
    # Second, Short Wall     
    # Pick
    prog.add_joint_move_to(tp_grip_set['j_pre_pick_2'], vel)
    prog.add_task_move_to(tp_grip_set['t_pick_2'], vel)
    close_gripper(prog)
    prog.add_task_move_to(tp_grip_set['t_pickup_2'], vel)
    
    
    # Place     
    prog.add_move_home()
    prog.add_joint_move_to(tp_grip_set['j_pre_place_2'], vel)
    prog.add_task_move_to(tp_grip_set['t_place_2-1'], vel)
    prog.add_task_move_to(tp_grip_set['t_place_2-2'], vel)
    prog.add_task_move_to(tp_grip_set['t_place_2-3'], vel)
    open_gripper(prog)
    
    # Reposition    
    prog.add_task_move_to(tp_grip_set['t_reposition_2-1'], vel)
    close_gripper(prog)
    prog.add_task_move_to(tp_grip_set['t_reposition_2-2'], vel)
    open_gripper(prog)
    prog.add_task_move_to(tp_grip_set['t_reposition_2-3'], vel)
    prog.add_move_home()
    
    # Third, Short Wall     
    # Pick
    prog.add_joint_move_to(tp_grip_set['j_pre_pick_3'], vel)
    prog.add_task_move_to(tp_grip_set['t_pick_3'], vel)
    close_gripper(prog)
    prog.add_task_move_to(tp_grip_set['t_pickup_3'], vel)
    
    
    # Place     
    prog.add_move_home()
    prog.add_joint_move_to(tp_grip_set['j_pre_place_3'], vel)
    prog.add_task_move_to(tp_grip_set['t_place_3-1'], vel)
    prog.add_task_move_to(tp_grip_set['t_place_3-2'], vel)
    prog.add_task_move_to(tp_grip_set['t_place_3-3'], vel)
    open_gripper(prog)
    
    # Reposition    
    prog.add_task_move_to(tp_grip_set['t_reposition_3-1'], vel)
    close_gripper(prog)
    prog.add_task_move_to(tp_grip_set['t_reposition_3-2'], vel)
    open_gripper(prog)
    prog.add_task_move_to(tp_grip_set['t_reposition_3-3'], vel)
    prog.add_move_home()
    
    
    
    # Forth, Long Wall
    # Pick
    prog.add_joint_move_to(tp_grip_set['j_pre_pick_4'], vel)
    prog.add_task_move_to(tp_grip_set['t_pick_4'], vel)
    close_gripper(prog)
    prog.add_task_move_to(tp_grip_set['t_pickup_4'], vel)
    
    # Place     
    prog.add_joint_move_to(tp_grip_set['j_pre_place_4'], vel)
    prog.add_task_move_to(tp_grip_set['t_place_4-1'], vel)
    prog.add_task_move_to(tp_grip_set['t_place_4-2'], vel)
    prog.add_task_move_to(tp_grip_set['t_place_4-3'], vel)
    open_gripper(prog)
    prog.add_task_move_to(tp_grip_set['t_post_place_4'], vel)
    prog.add_move_home()
    prog.add_joint_move_to(tp_grip_set['j_stay_back_1'], vel)
    prog.add_joint_move_to(tp_grip_set['j_stay_back_2'], vel)
   
            
    return json.dumps(prog.json_program)
   
    
```

```python
prog_gripper_setting_wall = indy_gripper_setting_wall()
indy_gripper.set_and_start_json_program(prog_gripper_setting_wall)
```
